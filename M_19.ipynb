```python
#Task 1
function Person(name, age) {
    this.name = name;
    this.age = age;
}

Person.prototype.displayInfo = function() {
    console.log(`Name: ${this.name}, Age: ${this.age}`);
};

const person1 = new Person("Alice", 25);
const person2 = new Person("Bob", 30);

person1.displayInfo(); // Output: Name: Alice, Age: 25
person2.displayInfo(); // Output: Name: Bob, Age: 30
```

```python
#Task 2
function BankAccount(initialBalance) {
    this.balance = initialBalance;
}

BankAccount.prototype.deposit = function(amount) {
    if (amount > 0) {
        this.balance += amount;
    }
};

BankAccount.prototype.withdraw = function(amount) {
    if (amount > 0 && amount <= this.balance) {
        this.balance -= amount;
    }
};

const account = new BankAccount(100);
console.log(`Initial balance: ${account.balance}`); // Output: Initial balance: 100
account.deposit(50);
console.log(`After deposit: ${account.balance}`); // Output: After deposit: 150
account.withdraw(30);
console.log(`After withdrawal: ${account.balance}`); // Output: After withdrawal: 120
```

```python
#Task 5
class Calculator {
    constructor(value1, value2) {
        this.value1 = value1;
        this.value2 = value2;
    }

    add() {
        return this.value1 + this.value2;
    }

    subtract() {
        return this.value1 - this.value2;
    }

    multiply() {
        return this.value1 * this.value2;
    }

    divide() {
        if (this.value2 !== 0) {
            return this.value1 / this.value2;
        } else {
            return "Cannot divide by zero";
        }
    }

    showOperations() {
        console.log(`Add: ${this.add()}`);
        console.log(`Subtract: ${this.subtract()}`);
        console.log(`Multiply: ${this.multiply()}`);
        console.log(`Divide: ${this.divide()}`);
    }
}

const calc = new Calculator(10, 5);
calc.showOperations();
// Output:
// Add: 15
// Subtract: 5
// Multiply: 50
// Divide: 2
```

```python
#task 6
class Shape {
    constructor() {
        if (new.target === Shape) {
            throw new Error("Shape is an abstract class and cannot be instantiated directly.");
        }
    }
}

class Circle extends Shape {
    constructor(radius) {
        super();
        this.radius = radius;
    }

    area() {
        return Math.PI * this.radius * this.radius;
    }

    perimeter() {
        return 2 * Math.PI * this.radius;
    }
}

class Rectangle extends Shape {
    constructor(width, height) {
        super();
        this.width = width;
        this.height = height;
    }

    area() {
        return this.width * this.height;
    }

    perimeter() {
        return 2 * (this.width + this.height);
    }
}

class RightTriangle extends Shape {
    constructor(base, height) {
        super();
        this.base = base;
        this.height = height;
    }

    area() {
        return 0.5 * this.base * this.height;
    }

    perimeter() {
        const hypotenuse = Math.sqrt(this.base * this.base + this.height * this.height);
        return this.base + this.height + hypotenuse;
    }
}

const circle = new Circle(5);
const rectangle = new Rectangle(4, 6);
const triangle = new RightTriangle(3, 4);

console.log(`Circle Area: ${circle.area()}, Perimeter: ${circle.perimeter()}`);
console.log(`Rectangle Area: ${rectangle.area()}, Perimeter: ${rectangle.perimeter()}`);
console.log(`Triangle Area: ${triangle.area()}, Perimeter: ${triangle.perimeter()}`);
// Output:
// Circle Area: 78.53981633974483, Perimeter: 31.41592653589793
// Rectangle Area: 24, Perimeter: 20
// Triangle Area: 6, Perimeter: 12.364898683664596
```

```python
#Task 7
class Book {
    constructor(title, author, availability = true) {
        this.title = title;
        this.author = author;
        this.availability = availability;
    }
}

class Library {
    constructor() {
        this.books = [];
    }

    addBook(book) {
        if (book instanceof Book) {
            this.books.push(book);
            console.log(`${book.title} by ${book.author} added to the library.`);
        } else {
            console.log("Invalid book object.");
        }
    }

    checkOutBook(title) {
        const book = this.books.find(b => b.title === title && b.availability);
        if (book) {
            book.availability = false;
            console.log(`${title} has been checked out.`);
        } else {
            console.log(`${title} is not available.`);
        }
    }

    returnBook(title) {
        const book = this.books.find(b => b.title === title && !b.availability);
        if (book) {
            book.availability = true;
            console.log(`${title} has been returned.`);
        } else {
            console.log(`${title} is already available or not found.`);
        }
    }

    getAvailableBooks() {
        const availableBooks = this.books.filter(b => b.availability);
        console.log("Available Books:");
        availableBooks.forEach(book => console.log(`${book.title} by ${book.author}`));
        return availableBooks;
    }
}

const library = new Library();
const book1 = new Book("The Great Gatsby", "F. Scott Fitzgerald");
const book2 = new Book("1984", "George Orwell");

library.addBook(book1);
library.addBook(book2);
library.checkOutBook("The Great Gatsby");
library.getAvailableBooks();
library.returnBook("The Great Gatsby");
library.getAvailableBooks();
// Output:
// The Great Gatsby by F. Scott Fitzgerald added to the library.
// 1984 by George Orwell added to the library.
// The Great Gatsby has been checked out.
// Available Books:
// 1984 by George Orwell
// The Great Gatsby has been returned.
// Available Books:
// The Great Gatsby by F. Scott Fitzgerald
// 1984 by George Orwell
```
